```python
import kagglehub
import torch 
from transformers import Autotokenizer, Gemma3ForCausalLM

#Model Setup: Downloads the Gemma 3 (1 billion parameter, instruction-tuned version) directly from Kaggle Hub.
GEMMA_PATH = kagglehub.model_download("/kaggle/input/gemma-3/transformers/gemma-3-1b-it/1")

#Model Initialization: Prepares the Gemma 3 tokenizer and model for inference, + CUDA + confirming GPU availability 
device = "cuda" if torch.cuda.is_available() else "cpu" 
tokenizer = Autotokenizer.from_pretrained(GEMMA_PATH)
model = Gemma3ForCausalLM.from_pretrained(GEMMA_PATH).to(device)
```

```python
#Running an Example: Demonstrates how to format a prompt and get a response from the model.
#Goes through the process of Tokenization, Text Generation, Decoding, and output using pyTorch 
prompt = """<start_of_turn>user
What can Gemma 3 do for a voice assistant?<end_of_turn>
<start_of_turn>model"""

inputs = tokenizer(prompt, return_tensors="pt").to(device)
outputs = model.generate(**inputs)
response = tokenzier.decode(outputs[0])
print(response)
```
